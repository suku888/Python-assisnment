# Theory

### 1. What is a random variable in probability theory?

##### A random variable is a variable whose value is determined by the outcome of a random phenomenon.

### 2. What are the types of random variables?

#####  i) Discrete: Takes countable values (e.g., number of students).

##### ii) Continuous: Takes any value in a range (e.g., height, weight).

### 3. What is the difference between discrete and continuous distributions?

#####  i) Discrete: Deals with distinct values.

##### ii) Continuous: Deals with data over a continuum.

### 4. What are probability distribution functions (PDF)?

##### A function that describes the likelihood of a random variable taking specific values (used for continuous variables).

### 5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?

#####  i) PDF: Shows probability at a single point.

##### ii) CDF: Shows cumulative probability up to a point.

### 6. What is a discrete uniform distribution?

##### Each value in a finite set has equal probability (e.g., rolling a fair die).

### 7. What are the key properties of a Bernoulli distribution?

#####  i) Only 2 outcomes: success (1) and failure (0).

##### ii) Probability of success = p.

### 8. What is the binomial distribution, and how is it used in probability?

##### It models the number of successes in a fixed number of independent trials (e.g., flipping a coin 10 times).

### 9. What is the Poisson distribution and where is it applied?

##### It models the number of events in a fixed interval (e.g., number of emails per hour).

### 10. What is a continuous uniform distribution?

##### Every value within an interval has equal probability (e.g., choosing a number between 0 and 1).

### 11. What are the characteristics of a normal distribution?

#####  i) Symmetrical bell shape.

##### ii) Mean = median = mode.

##### iii) Most values lie close to the mean.

### 12. What is the standard normal distribution, and why is it important?

##### A normal distribution with mean 0 and SD 1. Used in Z-score calculations and statistical tables.

### 13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?

##### CLT states that the sampling distribution of the mean becomes normal as the sample size increases, even if the population isn’t normal.

### 14. How does the Central Limit Theorem relate to the normal distribution?

##### It justifies the use of the normal distribution for inference even when the population is not normal.


### 15. What is the application of Z statistics in hypothesis testing?

##### Used to determine how far a sample mean is from the population mean under the null hypothesis.

### 16. How do you calculate a Z-score, and what does it represent?

##### It shows how many standard deviations a value is from the mean.

### 17. What are point estimates and interval estimates in statistics?

#####  i) Point estimate: Single value estimate (e.g., sample mean).

##### ii) Interval estimate: Range of values (confidence interval).

### 18. What is the significance of confidence intervals in statistical analysis?

##### It provides a range within which the true parameter is likely to lie with a certain confidence level (e.g., 95%).


### 19. What is the relationship between a Z-score and a confidence interval?

##### Z-scores are used to calculate the margin of error for confidence intervals.


### 20. How are Z-scores used to compare different distributions?

##### They standardize values so you can compare across different scales or distributions.

### 21. What are the assumptions for applying the Central Limit Theorem?

#####  i) Independent samples.

##### ii) Sample size is sufficiently large (n ≥ 30).

##### iii) Population with finite variance.

### 22. What is the concept of expected value in a probability distribution?

##### The long-run average or mean of a random variable.

E(X)=∑[x⋅P(x)]

### 1. Write a Python program to generate a random variable and display its value.

```python
import random

random_variable = random.randint(1, 100)
print("Random Variable:", random_variable)
```

### 2. Generate a discrete uniform distribution using Python and plot the probability mass function (PMF).

```python
import matplotlib.pyplot as plt
import numpy as np

values = np.arange(1, 7)
probabilities = [1/6] * 6

plt.bar(values, probabilities)
plt.title("Discrete Uniform Distribution (Die Roll)")
plt.xlabel("Value")
plt.ylabel("Probability")
plt.show()
```

### 3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

```python
from scipy.stats import bernoulli
import matplotlib.pyplot as plt

p = 0.6
x = [0, 1]
pdf = bernoulli.pmf(x, p)

plt.bar(x, pdf)
plt.title("Bernoulli Distribution PDF")
plt.xlabel("Outcome")
plt.ylabel("Probability")
plt.show()
```

### 4.Write a Python script to simulate a binomial distribution with n=10 and p=0.5, then plot its histogram.


```python
from scipy.stats import binom
import matplotlib.pyplot as plt

n = 10
p = 0.5
x = np.arange(0, n+1)
pmf = binom.pmf(x, n, p)

plt.bar(x, pmf)
plt.title("Binomial Distribution PMF (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Probability")
plt.show()
```

### 5. Create a Poisson distribution and visualize it using Python.

```python
from scipy.stats import poisson
import matplotlib.pyplot as plt

mu = 3
x = np.arange(0, 10)
pmf = poisson.pmf(x, mu)

plt.bar(x, pmf)
plt.title("Poisson Distribution (λ=3)")
plt.xlabel("Number of Events")
plt.ylabel("Probability")
plt.show()
```

#### 6. Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution.


```python
from scipy.stats import randint

x = np.arange(1, 7)
cdf = randint.cdf(x, 1, 7)

plt.step(x, cdf, where='mid')
plt.title("CDF of Discrete Uniform Distribution (1 to 6)")
plt.xlabel("Value")
plt.ylabel("Cumulative Probability")
plt.show()
```

### 10. Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.

```python
sample_means = []

for _ in range(1000):
    sample = np.random.exponential(scale=2.0, size=30)  # non-normal distribution
    sample_means.append(np.mean(sample))

plt.hist(sample_means, bins=30, edgecolor='black')
plt.title("CLT Demonstration with Exponential Distribution")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.show()
```

### 15. Simulate multiple samples from a normal distribution and verify the Central Limit Theorem.

```python
means = []
for _ in range(1000):
    sample = np.random.normal(loc=50, scale=10, size=30)
    means.append(np.mean(sample))

plt.hist(means, bins=30, edgecolor='black')
plt.title("CLT Verification: Sampling Distribution of Mean")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.show()
```

#### 16. Write a Python function to calculate and plot the standard normal distribution (mean = 0, std = 1).


```python
x = np.linspace(-4, 4, 1000)
pdf = (1 / np.sqrt(2 * np.pi)) * np.exp(-0.5 * x**2)

plt.plot(x, pdf)
plt.title("Standard Normal Distribution (mean=0, std=1)")
plt.xlabel("x")
plt.ylabel("PDF")
plt.grid(True)
plt.show()
```

#### 17. Generate random variables and calculate their corresponding probabilities using the binomial distribution.

```python
n = 10
p = 0.5
x = np.arange(0, n+1)
probs = binom.pmf(x, n, p)

for val, prob in zip(x, probs):
    print(f"P(X={val}) = {prob:.3f}")
```

 #### 18. Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution.

```python
def z_score(x, mean, std):
    return (x - mean) / std

print("Z-score:", z_score(55, 50, 10))
```

#### 19. Implement hypothesis testing using Z-statistics for a sample dataset.

```python
from scipy.stats import norm

mean, std, n = 50, 10, 30
sample_mean = 52
z = (sample_mean - mean) / (std / np.sqrt(n))
p_value = 1 - norm.cdf(z)

print(f"Z = {z:.2f}, P-value = {p_value:.4f}")
```

#### 20. Create a confidence interval for a dataset using Python and interpret the result.


```python
import scipy.stats as stats

data = np.random.normal(100, 15, 50)
mean = np.mean(data)
se = stats.sem(data)
ci = stats.t.interval(0.95, len(data)-1, loc=mean, scale=se)

print("95% Confidence Interval:", ci)
```

#### 21. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.

```python
data = np.random.normal(70, 10, 100)
mean = np.mean(data)
se = stats.sem(data)
ci = stats.t.interval(0.95, len(data)-1, loc=mean, scale=se)

print("Confidence Interval for Mean:", ci)
```

#### 25. Write a Python program to compare the standard deviations of two datasets and visualize the difference.

```python
data1 = np.random.normal(50, 10, 100)
data2 = np.random.normal(50, 20, 100)

std1 = np.std(data1)
std2 = np.std(data2)

plt.hist(data1, alpha=0.5, label=f"STD={std1:.1f}")
plt.hist(data2, alpha=0.5, label=f"STD={std2:.1f}")
plt.legend()
plt.title("Comparison of Standard Deviations")
plt.show()
```

#### 26. Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution.

```python
data = np.random.normal(50, 15, 100)
iqr = np.percentile(data, 75) - np.percentile(data, 25)
print("Range:", np.max(data) - np.min(data))
print("IQR:", iqr)
```
